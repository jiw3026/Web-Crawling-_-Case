특히 숫자와 쉽표가 함께 있는 이러한 경우 아래 예시와 같이 쉼표(,)로 인해 숫자형으로 쉽게 변환하지 못합니다.

우리는 추후 새로운 칼럼을 만들기 위하여 지금 데이터 타입을 옳게 바꿔주겠습니다.

```python
#예시: 이렇게 쉼표가 있으면 바로 형변환을 적용할 수 없습니다
int(1,409)
```

```python
# 기간과 자치구 칼럼을 제외하고 데이터 타입을 int로 변환
for col in seoul_crime.columns[2:]: #기간[0] ,자치구[1] 제외
    print(col)
    seoul_crime[col] = seoul_crime[col].astype(str) #.str은 string 값에만 적용할 수 있으므로 먼저 str로 일차변환 진행
    seoul_crime[col] = seoul_crime[col].str.replace(',', '').astype(int) # 쉼표 제거 후 int로 형변환
print('All done')
```

이제 각 범죄별 검거와 발생 건수를 포함하는 '검거율' 칼럼을 생성해보겠습니다.

※ 검거율 = 검거 수 /발생 수

```python
seoul_crime['합계 검거율'] = seoul_crime['합계 검거']/seoul_crime['합계 발생']*100
seoul_crime['살인 검거율'] = seoul_crime['살인 검거']/seoul_crime['살인 발생']*100
seoul_crime['강도 검거율'] = seoul_crime['강도 검거']/seoul_crime['강도 발생']*100
seoul_crime['강간/추행 검거율'] = seoul_crime['강간/추행 검거']/seoul_crime['강간/추행 발생']*100
seoul_crime['절도 검거율'] = seoul_crime['절도 검거']/seoul_crime['절도 발생']*100
seoul_crime['폭력 검거율'] = seoul_crime['폭력 검거']/seoul_crime['폭력 발생']*100

seoul_crime
```

```python
# 다시 인덱스를 해제합니다.
# 원본을 사용해도 괜찮으나 지금은 복사해서 사용하겠습니다.
crime_data = seoul_crime.reset_index()
crime_data
```

```python
# 합계 칼럼 삭제합니다
crime_data = crime_data[crime_data['자치구'] !='합계'] 
```

우리는 서울시의 구별 범죄 데이터를 통해 범죄의 발생 자체가 적거나 혹은

범죄가 발생했더라도 검거가 잘되는 일명 '안전한' 구가 어딘지 알아 볼 수 있습니다. (그 반대도 가능합니다) 

그럼 단순히 전체 범죄의 수로만 따졌을때 

가장 범죄가 많이 일어나는 서울시 구와 가장 범죄가 적게 일어나는 서울시 구를 각각 5개씩 찾아보겠습니다.

```python
years = ['2014','2015','2016','2017','2018','2019']

high ={}
low = {}

for year in years:
    data = crime_data[crime_data['기간'] == year]
    data= data.sort_values(by='총 범죄', ascending=False)

# 2014~2019 범죄가 많이 일어났던 서울시 top 5 구
    high_tmp = [gu for gu in data['자치구'][:5]]
    
    for ht in high_tmp:
        if ht not in high:
            high[ht] =1
        else:
            value = high[ht]
            high[ht] =value +1

# 2014~2019 범죄가 적게 일어났던 서울시 top 5 구
    low_tmp = [gu for gu in data['자치구'][-5:]]
    
    for lt in low_tmp:
        if lt not in low:
            low[lt] =1
        else:
            value = low[lt]
            low[lt] =value +1
```

```python
#딕셔너리에는 정렬을 제공하는 기본 함수가 없습니다.
#그러나 아래 opeator 라이브러리를 활용해 딕셔너리를 정렬할 수 있습니다. 
import operator

high= sorted(high.items(), key=operator.itemgetter(1), reverse =True)
low= sorted(low.items(), key=operator.itemgetter(1), reverse =True)

print(high)
print()
print(low)
```

```python
high5 = [h for h,_ in high[:5]]
low5 = [l for l,_ in low[:5]]

print('가장 범죄가 많이 일어나는 서울시 TOP 5구')
print(high5, end='\n\n')

print('가장 범죄가 적게 일어나는 서울시 TOP 5구')
print(low5)
```

이번에는 총 검거율로만 따졌을때 

가장 검거율이 높은 서울시 구와 가장 검거율이 낮은 서울시 구를 각각 5개씩 찾아보겠습니다.

```python
years = ['2014','2015','2016','2017','2018','2019']

weak ={}
strong = {}

for year in years:
    data = crime_data[crime_data['기간'] == year]
    data= data.sort_values(by='총 검거율', ascending=False)

# 2014~2019 범죄 검거율이 가장 낮은 서울시 top 5 구
    weak_tmp = [gu for gu in data['자치구'][-5:]]
    
    for wt in weak_tmp:
        if wt not in weak:
            weak[wt] =1
        else:
            value = weak[wt]
            weak[wt] =value +1

# 2014~2019 범죄 검거율이 가장 높은 서울시 top 5 구
    strong_tmp = [gu for gu in data['자치구'][:5]]
    
    for st in strong_tmp:
        if st not in strong:
            strong[st] =1
        else:
            value = strong[st]
            strong[st] =value +1
```

```python
#딕셔너리 정렬 
weak= sorted(weak.items(), key=operator.itemgetter(1), reverse =True)
strong= sorted(strong.items(), key=operator.itemgetter(1), reverse =True)

print(weak, end='\n\n')
print(strong)
```

```python
strong5 = [s for s,_ in strong[:5]]
weak5 = [w for w,_ in weak[:5]]

print('가장 검거율이 높은 서울시 TOP 5구')
print(strong5, end='\n\n')

print('가장 검거율이 낮은 서울시 TOP 5구')
print(weak5)
```

```python
print('범죄(수) 많은 곳:', high5)
print('검거율 높은 곳:',weak5, end='\n\n')

print('범죄(수) 적은 곳:',low5)
print('검거율 적은 곳:',strong5)
```

영등포구 와 마포구는 범죄수도 많지만 검거율도 높습니다.

반대로 도봉구는 범죄수도 적지만 검거율도 적습니다. 

동작구는 범죄수도 적고 검거율도 높습니다. 이 중 동작구가 가장 안전한 구라고 볼 수 있겠습니다.
