```
# 연결손익계산서 시트 불러오기
df_kakao = pd.read_excel('/content/drive/MyDrive/데이터 덕/DA-Project/기업동향분석_카카오/2020_3분기_카카오_실적발표.xlsx', sheet_name = 'IS(Con)' , header = 2)
df_kakao
```

```
# (단위: 백만원)열과 (Unnamed: 2)열을 합쳐서 인덱스로 사용
df_kakao['index'] = np.where(pd.notnull(df_kakao['(단위: 백만원)']) == True, df_kakao['(단위: 백만원)'], df_kakao['Unnamed: 2'])

# 필요없는 열 삭제
df_kakao = df_kakao.set_index('index').drop(['Unnamed: 0', '(단위: 백만원)', 'Unnamed: 2'], axis=1)

#엑셀 주석부분 삭제
df_kakao = df_kakao[:-1]

# 비어있는 행 삭제
df_kakao.dropna(axis=0, inplace=True)
df_kakao
```

```
# 각 카테고리 별 열이름 저장 
year = [2017, 2018, 2019, 2020]
qoq = df_kakao.columns[-4:-2]
yoy = df_kakao.columns[-2:]
quarters = df_kakao.columns[:-4].drop(year)
quarters
```

```
quarters = ['1Q19', '2Q19', '3Q19', '4Q19', '1Q20', '2Q20', '3Q20']

df_naver_quarter = df_naver[quarters]
df_kakao_quarter = df_kakao[quarters]
```

비교하고 싶은 내역들을 뽑아 데이터 프레임을 만들어봅시다

```
df_naver_quarter.index
```

```
df_kakao_quarter.index
```

```
# naver에는 영업이익률 행이 없으므로 만들어 주기 
df_naver_quarter.loc['영업이익률(%)'] = df_naver_quarter.loc['영업이익']*100 / df_naver_quarter.loc['영업수익']
df_kakao_quarter.loc['영업이익률(%)'] = df_kakao_quarter.loc['영업이익']*100 / df_kakao_quarter.loc['매출액']

# 비교하고 싶은 내역 뽑기 
df_naver_quarter = df_naver_quarter.loc[['영업수익', '서치플랫폼', '콘텐츠', '영업비용', '마케팅', '영업이익', '영업이익률(%)','당기순이익','조정 EBITDA']]
df_kakao_quarter = df_kakao_quarter.loc[['매출액', '플랫폼부문', '콘텐츠부문', '영업비용', '마케팅비', '영업이익', '영업이익률(%)','연결당기순이익','EBITDAR']]

# 칼럼명 통일 시키기 
df_naver_quarter.index = ['매출액', '플랫폼부문', '콘텐츠부문', '영업비용', '마케팅비', '영업이익', '영업이익률(%)','당기순이익','조정 EBITDA']
df_kakao_quarter.index = ['매출액', '플랫폼부문', '콘텐츠부문', '영업비용', '마케팅비', '영업이익', '영업이익률(%)','당기순이익','조정 EBITDA']


print('네이버 재무제표')
display(df_naver_quarter)

print('카카오 재무제표')
display(df_kakao_quarter)
```

```
#  두 데이터 프레임 합치기 
df_kakao_quarter['회사명'] = '카카오' # 회사이름 칼럼 추가 
df_kakao_quarter = df_kakao_quarter.reset_index() #인덱스 해제 

df_naver_quarter['회사명'] = '네이버'
df_naver_quarter = df_naver_quarter.reset_index()

df_compare = pd.concat([df_naver_quarter, df_kakao_quarter], axis = 0, ignore_index=True) # 두 데이터 프레임 합치기
df_compare
```
